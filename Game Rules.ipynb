# Game Description
Analog Tactics is a skirmish game where each player controls a group of units on a hexagonal grid. By securing objectives and eliminating enemy units, you gain victory points (VP). You win the game by being the first to gain 50 victory points.

Controlling Units is easy and intuitive. Units have points that they can use to move, attack, or use abilities. If you want to do something, just spend an appropriate attribute point and do that action. For instance, use an Attack point to attack, a Movement point to move, or a Special point to use an ability. Units also have attributes that tell you how much damage they do, how much health they have, and how easy it is to hit them.

This game focuses on a deep upgrade system unique to each class of units, where complexity and strategy comes mostly from how you upgrade and use your units over the course of the game. Each turn your units will level up to give you an opportunity to add a new ability through their pool of available abilities. Additionally, each class has a separate upgrade system dependent on your decisions throughout the game.

Use abilities and tactics to maximize the number of points you earn each round by either eliminating other players' units or capturing objectives.

# Game Flow Overview
Each round consists of each player taking their turn. During their turn, a player's units will all be placed on the game board with full health and attributes, allowing them to move, attack, and function in general to their fullest potential. Each player will spend their turn maneuvering their units to capture objectives and attacking their opponent's units to gain points and bonuses. After each person has taken a turn, the round ends and each player levels up their units by choosing new available abilities. Continue to gain points until the victory threshold has been reached.

# Mechanics Preview
- Unit attacks are resolved using D6 dice: the initiator's dice result plus their passive hit modifier compares to the target's dice result plus the target's passive evasion modifier. An attack always deals base damage. If the overall hit result is greater than the overall evasion result, the attack adds Critical Damage. 
- Damage is calculated from the attacker's total 'Damage' output minus the target's 'Armor' attribute. Resulting Damage is subtracted from the target's Health attribute. A unit that reaches zero health is removed from the board.
- Some abilities cost 'Special' points. Units have a certain number of Special points each round to spend on abilities
- Between player turns, players may use abilities that cost 'Reaction' points.
- All used points refresh to their maximum at the beginning of a player's turn.
- Units that have been eliminated respawn on the board at the beginning of a player's turn.
- Units use Line of Sight to find targets for ranged attacks.
- Objectives are captured by players and have a Health attribute. 'Decapturing' an objective means reducing the objective's health to zero. The objective then becomes neutral and may be captured by any other player.
- After each player has taken a turn, the round ends and each player levels up.

## Standard Objects on the Board

## Objectives
- Objectives are crucial to winning the game, as these are the primary way to gain VP
- Spaces immediately surrounding an objective function as a respawn area
- Each player begins the game controlling 1 objective
- Objectives have 10 Health. To decapture an objective, reduce its health to zero. The objective becomes neutral.
- Objectives have 1 Armor per adjacent friendly unit, and is immune to [Piercing](#Piercing)
- To capture a neutral objective, spend a point specified by your class.
- At the end of your turn, score VP equal to the number of objectives you control
- Objectives occupy spaces, so units normally cannot move through them

## Respawn Points
- Respawn Points are separate objects on the board that allow you to respawn units
- Capture a Respawn Point using an attribute point detailed by your class
- Respawn Points have no Health and can be freely captured by opposing players
- Units cannot move through Respawn Points
- The spaces adjacent to a Respawn Point you possess function as a Respawn Area
- Spaces adjacent to Respawn Points you own are considered adjacent for movement purposes.
- Each player possesses a permanent respawn point, so that there is not a situation where a player is completely prevented from placing their units on the board.

## Obstacles
- Obstacles are keyworded stationary objects that occupy 1 space
- By default Obstacles have 2 Armor and 1 Health
- Obstacles are immune to [Piercing](#Piercing)
- Separate Permanent Obstacles are placed at game setup and cannot be removed from the game board
- Obstacles cannot be placed 1 space from Objectives or Respawn points

# Core Game Mechanics

## Beginning A Game
- All players roll a D6. The highest player goes first, and all other players’ turns ordered by their rolls. If there is a tie, reroll until there is a winner.
- Choose a starting Objective and Respawn Point 
- Place starting units within your respawn areas
- Your units start the game with all abilities described in their class's tier 0

## Start of Round
- Before the round starts, you upgrade your character according to its level. Note all refreshed points.

|Things to do during your turn|
|:---|
| Use any available points in any order |
| Move any friendly unit |
| Attack any enemy unit within range |
| Use any available abilities |
| Capture Objectives or Respawn Points |

## Your Turn
- At the beginning of your turn, respawn all units next to a respawn or objective you own unless otherwise noted in your class
- You may use points from any unit in any order to move, attack, and use abilities.
- Each ability has a keyword at the end of it, showing what type of attribute point is used to use that ability. 
- Abilities with the ‘Passive’ keyword requires no points and is always active
- Abilities with the ‘Reaction’ keyword may be used during any player’s turn when the conditions of the ability are met
- If an ability can be used as either a Special or Reaction, a slash will be used at the end of the description (ex ‘Special/ Reaction’)
- If an ability costs more than one point, a plus sign will be used to indicate the combined points cost (ex. Attack + Special)

## Leveling Up
- Every round, automatically level up at the end of the round
- When you level up, you may pick any ability within the current ability tier list or any previous tier (Tier lists are labeled with abilities). For example, if tier 3 is level 8+ and you have just turned level 8, you may choose abilities from tier 3, tier 2, or tier 1.
- Leveling is different for each class; check class rules to determine what abilities you have access to
- Additionally, your stats increase per level, allowing you to do more things during your turn, increase your durability, or increase your damage output.

## Gaining Victory Points
- Gain 2VP per objective you control at the end of your turn
- Gain 1VP by eliminating an enemy Common
- Gain 3VP by eliminating an enemy Elite

## Winning the Game
- Obtain 50VP to end the game
- If a player reaches 50VP, finish out the round. The player who has the most VP over 50 wins.

### Unit Elimination
- Elimination occurs when a unit's Health is reduced to zero.
- Eliminated units are removed from the board (including Obstacles).

# Unit Descriptions

## Your Units
- Each unit is defined by its attributes and the values associated with its attributes
- A player takes their turn, using their units’ points to move, attack, and use abilities until they cannot do anything else, or voluntarily end their turn

|Expendable Attributes | Finite number of points per turn |
|:---|:---|
|Attack| Used for standard attacks |
|Movement| Used to move units or change direction |
|Special| Used for abilities during your turn |
|Reaction| Used for abilities any time when applicable |

| Constant Attributes | Modifies unit combat interaction |
|:---|:---|
|Base Damage| Amount of Damage dealt with a standard attack |
|Critical Damage| Amount of additional Damage dealt if Hit Roll exceeds Evasion Roll | 
|Armor| Amount of damage reduced from attacks |
|Evasion| Modifier added to an Evasion roll |
|Hit| Modifier added to a Hit roll |

##### Expendable Attribute: Attack, Movement, Special, Reaction
- Expendable attributes decrease every time you use an action point of that attribute type
- At the beginning of the round, all expendable attributes are refreshed

##### Constant Attributes: Damage, Armor, Evasion, Hit
- Constant attributes determine results of interactions and do not change when used

##### Attributes
- Health (HP) – When total damage taken reaches this number, your unit is eliminated and removed from the board
- Attack – Use 1 Attack point to deal damage equal to your Damage attribute
- Movement – Move a unit 1 space per point
- Special – Points used for Special abilities
- Reaction – Points used for Reaction abilities
- Base Damage – Amount of damage dealt from 1 standard Attack
- Critical Damage - Amount of additional damage added to a successful Attack Roll
- Armor – Subtract this number from an enemy attack’s Damage
- Evasion – Modifier added to any Evasion rolls
- Hit – Modifier added to any Hit rolls

## Health
- Health is subtracted whenever your unit is [damaged](#Damage)
- When Health reaches 0 the affected unit is removed from play and 'Eliminated'
- If a unit is Eliminated, remove it from the board
- Health does not refresh across rounds unless stated otherwise
- Leveling up affects the maximum Health of a unit automatically

## Attack
- All initiated Attacks are Range 1 (melee) unless otherwise stated
- All ranged Attacks (Range 2+) do not deal base damage unless otherwise stated
- Attack points are refreshed every round
- You may target enemy units within [Line of Sight](#Line of Sight). For range 1 attacks (1 space away) you may target units in the front 3 spaces (1,2,6). If you wish to target units outside of these spaces, you must use Movement/Reaction points to either change direction or move towards the enemy unit.
- Abilities may use 'Attack' as a keyword. If 'Attack' is mentioned in the ability, it is implied that an Evasion and Hit roll determines success of the attack.

## Movement
- Expend a Movement point to move a unit and/or change its Direction
- Movement points refresh after every round
- Any spaces with a unit in it is considered occupied
- Units cannot move through occupied spaces unless otherwise noted

![Sample Movement](Figure3.PNG)

### Example Movement: 
Unit U starts at position U with 2 movement points. U moves to U’ and spends 1 movement point, and has 1 left. U moves from U’ to U’’, spending the last movement point. U can no longer move because it has 0 movement points remaining. The unit will be able to move again next turn after refreshing its Movement points.

##### Directionality

- Units (U) have directionality depicted with an arrow. The front space is ‘1’, moving clockwise increasing to ‘6’

![Orientation](Figure2.PNG)

## Special
- Expend a Special point when using an ability with a Special cost
- Special points refresh after each round
- Some abilites are described as being an 'Attack'. Abilities that are 'Attacks' use attacking mechanics and may be reacted to as Attacks
- Some abilities cost a Special point and are described with 'on successful attack'. After the attack is resolved (determining Hit and Evasion), you may use the ability to modify your attack and apply its effect. 

## Reaction
- Expend a Reaction point to use a Reaction ability
- Reactions may be used whenever the conditions described by the reaction ability are met
- If no specific conditions are described in the ability, you can use that Reaction any time
- In general, Reactions can be used after (or sometimes before) a 'step' in the game has occurred. A 'step' includes taking any action by using a point, evaluating the effect of using a point, taking damage, losing Health, declaring an attack, etc.
- The same Reaction ability cannot be used multiple times per combat interaction
- Reaction points refresh after each round
- If multiple players want to react to the same game step, the unit with the highest evasion takes precedence. To further break ties, go in order of players starting with the current player's turn. (Example: Player 3's turn, Player 1 and 4 want to react but need a tie break. Player 4 will go next sooner than Player 1, and therefore reacts first).

## Base and Critical Damage
- Base Damage is the value associated with an initiated base attack
- Critical Damage is the value added if an Attack Roll is successful
- Certain abilities use a predetermined damage amount instead of your damage attributes
- 'Losing Health' is different than 'Taking Damage', as incoming damage is considered the precursor to losing health (after Armor/Reaction effects are considered).
- Dealing damage is not synonymous with a successful attack. An attack may be successful but still deal no damage due to damage reductions.

### Targeted Damage
- Damage that targets a unit is considered Targeted Damage
- Damage that selects spaces is not considered Targeted Damage

## Armor
- If a unit is damaged during an attack, take the incoming damage of the attacking unit and subtract the targeted unit's Armor value. This resulting value is subtracted from the targeted unit’s Health.
- Armor stays constant across rounds

## Evasion and Hit Modifier
- When initiating an attack or attacking ability, the initiating unit rolls a D6 ('rolling Hit'), then adds their Hit modifier. The targeted unit rolls a D6 ('rolling Evasion'), then adds their Evasion modifier. 
- If the initiating unit's total is higher, the attack is a success and damage is calculated and subtracted from the current health value of the targeted unit. 
- If the targeted unit's total is higher, the attack misses and no health is subtracted
- Ties result in an evasion, unless otherwise noted (equal hit total and evasion total results in an Evasion)
- Base Evasion and Hit modifiers are constant attributes
- If an ability defines the Hit of an attack (ex. 6 Hit), the targeted unit rolls Evasion against that value and the initiator of the attack does not roll or add modifiers. Similarly, if an ability defines evasion (ex. 6 Evasion), the unit initiating the attack rolls Hit against that value and the defending unit does not roll Evasion.

# Combat

## Resolving an Attack
- The initiating unit declares an attack targeting a unit within range. If nothing else is done, the damage would be calculated as the base damage of the initiator minus the armor of the target.
- The initiating player can optionally roll an Attack Roll. The initiaor would roll a D6 for Hit, adding their unit's Hit modifier.
- The targeted unit rolls a D6 for Evasion, adding their unit's Evasion modifier.
- If the targeted unit does not have line of sight of the initiating unit, the targeted unit must take a 3 as the default Evasion result.
- If the Hit Roll is higher than the Evasion Roll, add Critical Damage to the resulting damage. If Evasion is higher or the roll is equal, the attack misses and no damage occurs as a direct result of a Hit.

## Combat Hierarchy
- 1: Establish Initiating Unit and Targeted Unit. This costs Attack/Special points to initiate depending on Attack or Ability
- 2: Initiating Unit may also initiate an Attack Roll. Initiator rolls Hit, Targeted Unit rolls Evasion; Resolve. If Wounding, skip this step.
- 3: If targeted unit is hit (initiating unit has higher value), subtract Armor from incoming Damage. If Piercing, skip this step.
- 4: Targeted unit loses Health equal to final calculated damage.

### Advantage and Disadvantage
- Gaining advantage means you roll two D6 and take the higher.
- Gaining disadvantage means you roll two D6 and take the lower.

## Line of Sight
- A unit's line of sight begins in spaces 1,2, and 6 and includes all spaces extending from them 
- A unit has line of sight if an imaginary line between any corner of its occupied space can reach any corner of the targeted unit without passing through another occupied space

![Line of Sight](Figure4.PNG)

### Partial Line of Sight
- A unit has partial line of sight if the closest corner of the initating unit cannot reach the three closest corners of the target without passing through an occupied space. Only touching an edge of an occupied space does not block corners.
- If you target a unit within partial line of sight with an attack, you subtract 1 from your Hit for each occluded corner. Since you cannot target a fully occluded unit, you can only receive a penalty of 1 or 2 to your Hit.
- Partial line of sight counts as full line of sight for ability purposes unless specifically mentioned

![Partial Line of Sight](Partial Line of Sight.PNG)

- Examples of partial line of sight. Black hexes are blocked spaces. Red hexes are targetable units with partial line of sight.

## Ranged Attacks
- Ranged Attacks can target units multiple spaces away (more than 1). The range value determines how many spaces away units can be targeted.
- Ranged Attacks cannot target spaces out of line of sight unless otherwise noted.

## Ranged Abilities
- Range 1 abilities are synonymous with a melee attack (1 space away).
- Ranged abilites 'within' a number of spaces include all spaces at a shorter distance.
- Ranged abilites 'at' a certain range are specifically at that range only.

### Nondirectional
- Line of Sight rules do not apply. Evaluate a normal combat interaction.

![Sample Combat](Figure6.PNG)

### Sample Combat: 
Unit A is facing northeast and wishes to attack unit B. Unit A has 2 attack points remaining, a hit modifier of 0, 1 Base Damage and 1 Critical Damage. Because B is in unit A’s #2 space, it is within line of sight and can therefore be targeted. Unit B is facing southeast, has an Evasion modifier of 2, has 1 Armor, and 1 Health left. Because unit B does not have line of sight of unit A, unit B must take 3 by default. Unit A needs to succeed on a Critical Roll to eliminate the target, since the base damage and armor would cancel out. Unit A rolls a 4 for Hit. Adding Hit and Evasion modifiers respectively, Hit and Evasion are both 5, and ties go to the defender. Unit A deals 1 damage, minus 1 armor from Unit B. No Health changes happen to B, and A has 1 less Attack point. Unit A attacks again. This time player A rolls a 5. Adding modifiers, 6 Hit is greater than 5 Evasion. Unit B subtracts 2 damage – 1 armor from its Health. Unit B is at 0 health and is eliminated and removed from the board. Player A gains VP for eliminating the target. If Unit B had line of sight of Unit A, both players would roll a D6 to add to their modifiers.

## Wounding
- This keyword denotes adding Critical Damage automatically with no Attack roll
- Armor is still applied when calculating reduction of Health

## Piercing
- This keyword denotes damage that ignores Armor
- Does not apply to Obstacles or Objectives

## Swift
- This keyword denotes attacks or damaging abilities that cost two reaction points to react to
- This applies to Reactions that would alter the normal combat interaction in any way

## Area of Effect (AOE)
- Abilities denoted with AOE deal damage to all surrounding spaces (1-6) from the target space. If no damage is specified, the base damage of the initiator is the default roll. Roll one Hit roll and apply to all targets when rolling for Critical Damage. Usually, rolling for Critical Damage is optional, here it is not.
- By default AOE damage does not affect friendly units, although abilities may affect friendly units if specified

## Forced Movement
- Some abilities cause other units to move into other spaces.
- If an ability says 'move away' the numbers of spaces must increase between the origin and the unit, as opposed to moving a unit sideways into another space that is the same distance away.
- If an ability says to 'move towards', same rules apply but the distance must decrease.
- If an ability says to move 'in a straight line' your unit must move the entire distance in one direction without changing direction.
- If a unit can move through occupied spaces, that unit moves through spaces with forced movement.
- Obstacles cannot be moved unless stated otherwise.

## Damage from Blocked Movement
- For any ability that forces a unit to move, if the space that the unit would move into is occupied, deal Damage to that unit and the unit in the occupied space instead of moving the unit. The damage is equal to half of the blocked Movement rounded up.
- If the occupied space blocking movement has an Obstacle, disregard damage to the Obstacle
- Normal damage interactions from blocked movement is not considered with units that can move through occupied spaces
- If the unit can move through occupied spaces, but would land in an occupied space, place the moved unit in the closest unoccupied space

## Resolving Abilities Affecting Multiple Units
- If an ability affects more than one unit at once, resolve the ability in player order, first with the Elite, then with Commons in ascending order

The Warrior is based around high damage, durable units that use attacks for their main damage output. By using different weapons at different times, your attacks have additional effects. The upgrade system is based around upgrading your weapons through trial by combat.

|Unique Mechanics| |
|:---|:---|
|Weapons| Use as Attack Modifiers |
|Weaponform| What effect your weapon will have |

- Commons respawn automatically at the beginning of your turn.

- [Weapon](#Weapons): You must declare a weapon to be used with any Attacks. A Weapon acts as an attack modifier and requires no additional points. Only one weapon at a time can be used with an Attack point. 
- Weaponform: Each Attack begins at Weaponform 1. Each Attack with a Weapon during your turn increases your Weaponform by 1. Maximum Weaponform is 4. You may use different Weapons for each Weaponform number. Increasing Weaponform at 4 returns to Weaponform 1. Decreasing Weaponform at 1 stays at 1.
- Each weapon has a corresponding ability with the unit's current Weaponform number. You must use the ability corresponding with your Weaponform number. For example, if your Weaponform is currently 2, you may not use any Form 1’s or Form 3’s, etc.
- Weaponform persists across rounds. Weaponform resets after the unit is eliminated.
- If you chose to not use a weapon, evaluate a standard Attack and your weaponform does not increase.
- All units have separate weapons, weapon upgrades, and weaponforms
- Every time your Common eliminates an enemy unit or decaptures an objective, upgrade one weapon on that Common.
- Every time your Elite eliminates an enemy unit or decaptures an objective, upgrade one weapon on all of your units.
- Each of your units upgrades one weapon once for your first 4 turns
- Abilities with the weapon named in front of it (i.e. Spear: Ability. Special) can only be used with the corresponding Weapon
- If there is no Weapon specifier in front of the upgrade, the upgrade applies in general, not only when using those Weapons. For example, gaining +2 Hit passive from Rapier+3 applies to all attacks.
- Movements do not require a Movement point unless you are uses a 'Movement' action. All bonus movement from weapon effects are free.

- Keyword: Sequence. A Sequence keyword designates a sequence of actions that are taken by using the associated cost (usually 1 Attack or 1 Movement point). 
- 'Facing' a space causes a unit to change [directionality](#Directionality). 
- An example ability with the Sequence keyword: "Attack space 1, Move into space 2, Face space 3. Sequence. Attack." The unit will Attack space 1, move into space 2 (keeping the same directionality), then face the unit's new space 3. The Sequence keyword is listed, then the cost of 1 Attack point.
- Units do not change directionality if directed to 'Move' into a numbered space during a Sequence.
- If your unit cannot complete a 'Move' action in a sequence, end the Sequence.
- If your unit cannot target a unit with an 'Attack' action, skip the Attack. The Sequence does not end.
- If no space is specified, your unit may target any unit within [line of sight](#Line of Sight)

## Assassin
-----------
The Assassin relies on your Commons to increase your Elite's damage output for high burst damage. High Damage and Evasion makes the Assassin a slippery character that can strike at any moment. The upgrade system is based around eliminating Elites and Objectives to rise the ranks in the Houses.

|Unique Mechanics| |
|:---|:---|
| Attack Bonus | Your Elite gains damage when your Commons attack enemies |
| Stealth Tokens | Place on the board to block enemy line of sight |
| Houses | Upgrade your House to gain additional abilities |

- If a Common successfully damages an enemy, increase the Elite Critical by 1. The Common dealing damage must be out of Line of Sight of any units on the targeted unit's team. Max Damage bonus is 15.
- If you eliminate an enemy Elite, reset Elite Damage bonus.
- If you eliminate an enemy Common or decapture an Objective, lose damage bonus by amount of Health lost by the target unit in that combat interaction.
- Any increase to Elite Critical occurs before a reset; i.e. there is no residual damage bonus after eliminating an Elite.

- The Assassin can obtain upgrades through each of the [Houses](#Houses).
- Whenever you eliminate an Elite, you increase the rank within one House of your choice. With increased rank, your Elite and Commons gain corresponding abilities corresponding to their rank in the House.
- You may also increase a House rank if you decapture an objective.
- Your Elite and Commons have the abilities defined by the House upgrades.
- House upgrades immediately go into effect, as opposed to the end of the round.
- At the end of the first round, you gain 1 upgrade in a House.

- Capturing Objectives: Spend an attack point to capture an adjacent neutral [Objective](#Objectives) or respawn point

- Assassin Ranks: Novice; Resident; Seneschal; Vizier; Grand Master
- Recruit: Alternative Recruits are placed on the board next to any available respawn area

##### Stealth Tokens: 
- A stealth token is a attribute-less token that occupies a space and blocks line of sight for enemy units but not your units. 
- Stealth tokens cannot be placed in any already occupied space.
- Any unit may move into spaces occupied by a stealth token.
- Stealth tokens are removed from the board by dealing any damage to the space they occupy.
- At the beginning of your turn, you may keep the number of Stealth Tokens on the board equal to your Elite damage bonus. Remove any excess stealth tokens at the beginning of your turn.

## Mage
-----------
The Mage is a highly flexible class that is based on using Special points to deal ranged damage. The capabilities and utility of your units depend on which Elements you attune throughout the game. The upgrade system is turn based, increasing your Elemental use the more you attune them. 

|Unique Mechanics| |
|:---|:---|
|Cooldown| Useable once per turn |
|Elements| Determines the abilities you can use and what attributes are upgraded |
|Attunement| Determines Element and some ability powers |

- Keyword: Cooldown: This ability can only be used once per turn.
- Every round at the beginning of your turn, choose one Element to attune. This increases the Attunement of that Element and gives you access to all abilities of that element up to your current level. 
- The Attunement for each Element also allows you to use abilites from that element when you are not attuned. You may use up to the level of abilities equal to your Attunement.
- For example, if your Air Attunement is 2, you have access to the level 1 and level 2 Air abilities when attuned to an element different than Air.
- Any reference to Attunement in abilities reference the Attunement of the element whose skill tree it belongs, and not to the Attunement of the current element you are attuned to.

- You always have access to Tier 0 abilities, regardless of which element you are attuned to
- At the end of the round when you gain levels, your Commons gain stat bonuses according to the new level and Element you are currently attuned to.

## Engineer
--------------
The Engineer is a flexible class where your units can be individually upgraded with Blueprints to form a unique combination of Damage output and utility. Gameplay focuses around building your lattice of defense and aura-based Obstacles to achieve map control. Highly durable Commons and a weak Elite.

|Unique Mechanics| |
|:---|:---|
|Build| Unit or Obstacle appears in an empty adjacent space to your Elite |
|Blueprints| Allocate to your Commons when built to customize |

- Commons do not automatically respawn and must be built. Built Commons may respawn within 1 space of your Elite
- If you eliminate an enemy unit, add a Blueprint to the your Blueprints pool. When building Commons, you may add upgrades to it by allocating the requisite amount of Blueprints. Any eliminated Common returns Blueprints to your pool. You may not stack identical Blueprints.
- You begin the game with 3 Blueprints. You may allocate these to your Commons at the beginning of the game.
- The Engineer Elite may move through any friendly built units
- All of your built units/obstacles count towards increasing an adjacent friendly Objective's armor

- Capturing Objectives: Spend a Special point to capture an adjacent neutral [objective](#Objectives) or respawn point
- Decapturing Objectives gives a Blueprint
- When your Elite is eliminated, gain 1 Blueprint.
