# Overview
Analog Tactics is a game where each player controls a group of units. By securing objectives and eliminating enemy units, you gain victory points (VP) and experience points (XP) to level up your units. You win the game by being the first to gain 15 victory points.

## Units and Classes
- A player controls their team of units, consisting of 1 elite unit and 4* common units
- Warrior: Abilities are tied to attacks. Strategically use weapons to maximize your damage output
- Assassin: Power up your elite by attacking with your commons. Be aware of position and line of sight
- Engineer: Power level is determined by position relative to your buildings. 
- Mage: Flexible class with heavy ability usage. 

## The Board
- The game board consists of hexagonal spaces. 
- Each space is described as a ‘space’ in relevant abilities
- On the sample board, red spaces designate objectives, and blue spaces designate permanent obstacles. Objective and obstacle placements can change to suit players’ preferences.

## Game Flow

- The game consists of round and turns.
- A round begins before the first player takes their turn. Use the time before the first player takes their turn to decide unit upgrades and note refreshed points
- At the beginning of a round, all units refresh their expendable attribute points
- Units respawn at the beginning of that player’s turn unless stated otherwise
- At the beginning of a round, all units refresh their expendable attribute points
- Any dice rolls use a six-sided die, and are referred to as a D6

## Respawning Units
- At the beginning of your turn, you may respawn a unit on any space not within 1 space of an enemy unit.
- If no space exists, you may place your unit on any edge space
- Each class has its own respawn rules

## Your Units
- Each unit is defined by the following attributes
- Each attribute has numbers associated with it
- A player takes their turn, using their units’ points to move, attack, and use abilities until they cannot do anything else, or voluntarily ends their turn

##### Expendable Attribute: Attack, Movement, Special, Reaction
- Expendable attributes consist of points and decrease every time you use an action of that attribute type

##### Constant Attributes: Damage, Armor, Evasion, Hit
- Constant attributes do not change when used

##### Attributes
- Health (HP) – If this number reaches 0, your unit is eliminated
- Attack – Use 1 attack point to deal damage equal to your damage attribute
- Movement – Move a unit 1 space per point
- Special – Points used for special abilities
- Reaction – Points used for reaction abilities
- Damage – Amount of health taken away from 1 standard attack
- Armor – Subtract this number from an enemy attack’s damage
- Evasion – Modifier added to any evasion rolls
- Hit – Modifier added to any attack rolls

## Health
- Health is subtracted whenever your unit is damaged
- When health reaches 0, the affected unit is removed from play
- Health is not affected across rounds unless stated otherwise
- Levelling up affects the maximum health of a unit, but does not immediately affect a unit’s current health, i.e. if a unit currently has 5 HP and levels up, any increases to maximum health do not raise the unit’s HP from 5 to 6 at the beginning of the next round

## Attacks
- All initiated attacks are range 1 unless otherwise stated
- Your attack gains +1 hit if initiated from the target’s #4 space (behind the target)
- Your target gains +1 evasion if initiated from the target’s #1 space (in front of target)
- Attack points are refreshed every round
- You may target enemy units within line of sight. This means for range 1 attacks (1 space away) you may target units in the front 3 spaces (1,2,6). If you wish to target units outside of these spaces, you must use movement points to either change orientation or move towards the enemy unit.

## Movement
- Expend a movement point for each a unit moves unless stated otherwise.
- Movement points refresh after every round
- You may change your unit's orientation after every movement
- Any spaces with a unit on it is considered occupied
- Units cannot move through occupied spaces unless otherwise noted

![Sample Movement](Figure3.png)

### Example Movement: 
Unit U starts at position U with 2 movement points. U moves to U’ and spends 1 movement point, and has 1 left. U moves from U’ to U’’, spending the last movement point. U can no longer move because it has 0 movement points remaining. The unit will be able to move again next turn after refreshing its movement points.

## Special
- Expend a special point when using an ability with a Special cost
- Special points refresh after each round
- Some abilites are described as being an 'attack'. Abilities that are 'attacks' use attacking mechanics and may be reacted to as attacks

## Reaction
- Expend a reaction point when using a reaction
- Reactions may be used whenever the conditions described by the reaction ability are met
- Reaction points refresh after each round

## Damage
- This is the value associated with an initiated base attack. 
- Damage stays constant across rounds and is subject to change from levelling up or abilities

## Armor
- If a unit is damaged during an attack, take the damage attribute of the attacking unit and subtract the targeted unit's armor value. This value is subtracted from the targeted unit’s health
- Armor stays constant across rounds

## Evasion and Hit Modifier
- When initiating an attack or damaging ability, the initiating unit rolls a D6 (rolling hit), then adds their hit modifier. The targeted unit rolls a D6 (rolling evasion), then adds their evasion modifier. 
- If the initiating unit rolls higher, the attack is a success and damage is calculated and subtracted from the current health value of the targeted unit. 
- If the targeted unit rolls higher, the attack misses and no health is subtracted
- Ties go to the targeted unit, unless otherwise noted
- Base evasion and hit modifiers are constant across rounds
- If an ability defines the hit of an attack (ex. 6 hit), the targeted unit rolls evasion against that value and the initiator of the attack does not roll or add modifiers. Similarly, if an ability defines evasion (ex. 6 evasion), the unit initiating the attack rolls hit against that value and the defending unit does not roll evasion.

## Leveling Up
- Every time your unit eliminates an enemy unit, you gain 1XP
- When you gain enough XP (1 or 2 per level), you level up. Levelling up is detailed by your specific class. 
- When you level up, you may pick any ability within the current ability tier list or any previous tier (Tier lists are labeled with abilities). For example, if tier 3 is level 8+ and you have just turned level 8, you may choose abilities from tier 3, tier 2, or tier 1.
- Additionally, your stats increase per level, allowing you to do more things during your turn
- You may only level up twice per turn. Any additional XP is converted to VP.

### Automatic Leveling
- At the end of the round after applying gained XP (leveling up), if you are below level 5, you automatically gain 1 level

## Gaining Victory Points
- Obtain 15VP to win the game
- Gain 1VP by eliminating an enemy elite
- Gain 1VP by securing an objective
- Gain VP by gaining additional XP past the requirements to level up twice during a single turn

## Line of Sight
- A unit's line of sight begins in spaces 1,2, and 6 and includes all spaces extending from them 
- Line of sight is considered blocked if a straight line from the center of the origin space to the center of the target space touches an occupied space

![Line of Sight](Figure4.PNG)

## Ranged Attacks
- Ranged attacks can target units multiple spaces away. The range value determines how many spaces away units can be targeted
- Ranged attacks cannot target spaces out of line of sight

![Ranged Attack](Figure5.PNG)

- Units can take advantage of +1 hit from space 4 if your range can travel through space 4
- Example: T and N have range 3. T gains +1 hit on its attack. N does not
- Units can avoid disadvantage of +1 evasion for target if your range can avoid space 1

![Sample Combat](Figure6.PNG)

### Sample Combat: 
Unit A is facing northeast and wishes to attack unit B. Unit A has 2 attack points remaining, a hit modifier of 0, and 2 damage. Because B is in unit A’s #2 space, it is within line of sight and can therefore be targeted. Unit B is facing southeast, has an evasion modifier of 2, has 1 armor, and 1 health left. Because unit A is targeting unit B from B’s #4 space, unit A gains +1 hit modifier. Both players roll a D6. Unit A rolls a 3, and unit B rolls a 2. Adding hit and evasion modifiers respectively, hit and evasion are both 4, and ties go to the defender. No changes happen to B, and A has 1 less attack point. Unit A attacks again. This time player A rolls a 5, and player B rolls a 2. Adding modifiers, 6 hit is greater than 4 evasion. Unit B subtracts 2 damage – 1 armor from its health. Unit B is at 0 health and is eliminated and removed from the board. Player A gains XP and VP if applicable. 

## Objectives

- End your turn with a unit on an objective to gain the 1XP or 1VP associated with that active objective
- A central objective is always active 
- Objective 1 begins as an active objective
- Objectives rotate every 2 turns
- At the start of the game, the central objective is active with 1 XP, and objective 1 is active with 1VP. The next round, the central objective is active with 1 VP, and objective 1 is active with 1 XP. Afterwards, the objectives rotate with the central objective active with 1 XP and objective 2 active with 1VP, and so on

## Beginning A Game
- All players roll a D6. The highest player goes first, and all other players’ turns ordered by their rolls. If there is a tie, reroll until there is a winner.
- Take turns placing units on the board until all units are placed
- Take turns placing objective and permanent obstacle locations if you want a custom board
- Your units start the game with all abilities described in their class's tier 0

## Your Turn
- At the beginning of your turn, respawn all units unless otherwise stated in the class description
- You may use points from any unit in any order to move, attack, and use abilities.
- Each ability has a keyword at the end of it, showing what type of attribute point is used to use that ability. 
- Abilities with the ‘Passive’ keyword requires no points and is always active
- Abilities with the ‘Reaction’ keyword may be used during any player’s turn when the conditions of the ability are met
- If an ability can be used as either a special or reaction, a comma will be used at the end of the description (ex ‘Special, Reaction’)
- If an ability costs two points, a plus sign will be used (ex. Attack + Special)

## Warrior
-------------
- Weapon: You may declare a weapon to be used with an attack. Acts as an attack modifier and requires no additional points. Only one weapon at a time can be used with an attack point. 
- Weaponform: Each attack point begins at weaponform 1. Each attack with a weapon during your turn increases your weaponform by 1. Maximum weaponform is 4. You may use different weapons for each weaponform number. 
- Each weapon has a corresponding ability with your current weaponform number. You must use the ability corresponding with your weaponform. For example, if your weaponform is currently 2, you may not use any form 1’s or form 3’s, etc.
- If you chose to not use a weapon, use do a standard attack and your weaponform does not increase.
- Eliminating an enemy unit upgrades the weapon used to eliminate that unit. For example, your unit uses Warhammer when attacking to eliminate a unit. Your unit's Warhammer becomes Warhammer+1, and you gain the upgrade associated with Warhammer+1. Weapons upgraded this way is unit specific (i.e. Warhammer+1 only applies to the specific unit that eliminated an enemy with Warhammer). This allows your units to upgrade different weapons and specialize with different abilities and builds.
- Commons may only use up to weaponform 3 and may only upgrade a weapon up to +2.
- Reactions gained by upgrading weapons may only be used if you last used that weapon with that unit. For example, assume your unit has Reaction abilities associated with Spear and Katana. If your unit used the Spear last, you may use Reactions given by your Spear upgrades, but not from your Katana upgrades.
- Passive abilities gained from weapon upgrades apply directly to your unit, and therefore can be used with other weapons. For example, gaining +2 hit passive from Rapier+3 applies to all attacks from all weapons.
